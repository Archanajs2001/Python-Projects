##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt
x = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
y = 22.2, 17.6, 8.8, 8, 7.7, 6.7
plt.bar(x,y,color= 'b', ec='k')
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Language\nWorldwide Oct 2017,compared to a year ago')
plt.grid(color='r', which= 'major', lw='0.5')
plt.grid(which= 'minor', linestyle= ':', color= 'gray', lw='0.5')
plt.minorticks_on()

plt.show()
````

##Exercise 1: Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties: 
```

Y label name = Total profit
```

The line plot graph should look like this.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('company_sales_data.csv')

sns.lineplot(data,x='month_number',y='total_profit')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.xticks(range(1,13))
plt.yticks([100000,200000,300000,400000,500000])
plt.show()
````

##Exercise 3: Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

The graph should look like this.



image.png

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('company_sales_data.csv')
df = pd.DataFrame(data)

monthnumber = df['month_number'].tolist()
facecream = df['facecream'].tolist()
facewash = df['facewash']
toothpaste = df['toothpaste'].tolist()
bathingsoap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
moisturizer = df['moisturizer']

plt.plot(monthnumber,facecream, marker='o',label='Face cream Sales data', lw=3)
plt.plot(monthnumber,facewash, marker='o', label= 'Face wash Sales data',lw=3)
plt.plot(monthnumber,toothpaste,marker='o', label= 'Toothpaste Sales data',lw=3)
plt.plot(monthnumber,bathingsoap,marker='o', label= 'Bathing soap Sales data',lw=3)
plt.plot(monthnumber,shampoo,marker='o', label= 'Shampoo Sales Data',lw=3)
plt.plot(monthnumber,moisturizer,marker='o', label= 'Moisturizer Sales data',lw=3)

plt.legend()
plt.xlabel('Month number')
plt.ylabel('Sold units in number')
plt.title('Sales data')
plt.xticks(monthnumber)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])

plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('company_sales_data.csv')
df = pd.DataFrame(data)
facecream = df['facecream'].tolist()
facewash = df['facewash']
toothpaste = df['toothpaste'].tolist()
bathingsoap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
moisturizer = df['moisturizer']

x = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]

plt.pie(x, labels=['Face cream','Face wash', 'Toothpaste', 'Bathing soap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales Data')
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('company_sales_data.csv')
df = pd.DataFrame(data)

monthnumber = df['month_number'].tolist()
facecream = df['facecream'].tolist()
facewash = df['facewash']
toothpaste = df['toothpaste'].tolist()
bathingsoap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
moisturizer = df['moisturizer']
l = ['Face cream','Face wash','Toothpaste','Bathing soap','Shampoo','Moisturizer']


plt.stackplot(monthnumber, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer, colors=['m','c','r','k','g','y'], labels= l)
plt.legend(loc= 'upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('All product Sale data using Stack plot')

plt.show()
````
